# 01 Data Visualization

# 0. Prerequisites

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import pandas as pd
import numpy as np
```

# 1. Matplotlib for creating plots

## 1.1 Line charts

```python
# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line chart
plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## 1.2 Bar charts

```python
# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 4]

# Create a bar chart
plt.bar(categories, values)
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

## 1.3 Scatter plots

```python
# Sample data
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## 1.4 Customizing plot styles and aesthetics

```python
# Customizing plot styles
plt.style.use('ggplot')

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line chart with customized style
plt.plot(x, y, label='Sine Wave', color='blue', linestyle='--', marker='o')
plt.title('Styled Sine Wave')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

# 2. Seaborn for statistical visualization

## 2.1 Box plots

```python
# Sample data
data = np.random.normal(size=100)

# Create a box plot
sns.boxplot(data=data)
plt.title('Box Plot Example')
plt.show()
```

## 2.2 Violin plots

```python
# Sample data
data = np.random.normal(size=100)

# Create a violin plot
sns.violinplot(data=data)
plt.title('Violin Plot Example')
plt.show()
```

## 2.3 Heatmaps

```python
# Sample data
data = np.random.rand(10, 12)

# Create a heatmap
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Heatmap Example')
plt.show()
```

## 2.4 Pair plots for correlation analysis

```python
# Sample data
iris = sns.load_dataset('iris')

# Create a pair plot
sns.pairplot(iris, hue='species')
plt.show()
```

# 3. Interactive visualizations with Plotly

## 3.1 Creating interactive charts and dashboards

```python
# Sample data
df = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'size': np.random.rand(50) * 100,
    'color': np.random.rand(50)
})

# Create an interactive scatter plot
fig = px.scatter(df, x='x', y='y', size='size', color='color', title='Interactive Scatter Plot')
fig.show()
```

## 3.2 Adding interactivity and tooltips

```python
# Sample data
df = pd.DataFrame({
    'x': np.random.rand(50),
    'y': np.random.rand(50),
    'size': np.random.rand(50) * 100,
    'color': np.random.rand(50),
    'label': [f'Point {i}' for i in range(50)]
})

# Create an interactive scatter plot with tooltips
fig = px.scatter(df, x='x', y='y', size='size', color='color', hover_name='label', title='Interactive Scatter Plot with Tooltips')
fig.show()
```
