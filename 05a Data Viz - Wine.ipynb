```python
df['normalized_free'] =  (df['free sulfur dioxide'] - df['free sulfur dioxide'].mean())/df['free sulfur dioxide'].std()
df['normalized_free'].plot(kind='box')
```

```python
def category(input_):
    if input_ < 5:
        return 'L'
    elif (input_ >= 5) & (input_ < 7):
        return 'M'
    elif input_ >= 7:
        return 'H'
    else:
        return 'N/A'
```

```python
df['Quality Category'] = df['quality'].apply(category)
df['Quality Category'].value_counts()
```

```python
sns.pairplot(df[['alcohol','density']], hue='Quality Category')
```

```python
# df[df['Quality Category'] == 'M'][['volatile acidity']]
plt.scatter(df[df['Quality Category'] == 'H'][['volatile acidity']], df[df['Quality Category'] == 'H'][['citric acid']], alpha=0.3, color='green')
plt.scatter(df[df['Quality Category'] == 'M'][['volatile acidity']], df[df['Quality Category'] == 'M'][['citric acid']], alpha=0.1, color='red')
plt.scatter(df[df['Quality Category'] == 'L'][['volatile acidity']], df[df['Quality Category'] == 'L'][['citric acid']], alpha=0.3, color='blue')
```
